Q1 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_even_numbers(numbers):
  return sum(num for num in numbers if num % 2 == 0)
numbers = list(map(int,input("Enter the numbers: ").split()))
result = sum_even_numbers(numbers)
print("Sum of even numbers:", result)
```

Q3 Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.

```python
l = [1, 2, 3, 4, 5]

def square(l):
    sq = []
    for i in l:
        sq.append(i**2)
    return sq
```

```python
square(l)
```

Q4 Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
  if n<1 or n>200:
    return None
  for i in range(2,n):
    if n%i==0:
      return False
  return True

num = int(input("Enter a number in the range of 1 to 200: "))
result = is_prime(num)
if result is None:
  print("Number is out of range.")
elif result:
  print(f"{num} is a prime number.")
else:
  print(f"{num} is not a prime number.")
```

Q5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
def fib(n):
  a = 0
  b = 1
  for i in range(n):
    yield a
    a, b = b, a+b

n = 7
f = fib(n)
for i in range(n):
    print(next(f))
```

Q6 Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def power_of_2(n):
    for i in range(n):
        yield 2 ** i
n = 5
for i in power_of_2(n):
    print(i)
```

Q7 Implement a generator function that reads a file line by line and yields each line as a string.

```python
def create_file(file_path):
    content = """This is the first line
This is the second line
This is the third line"""
    with open(file_path, 'w') as file:
        file.write(content)
file_path = 'example.txt'
create_file(file_path)

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

file_path = 'example.txt'
for line in read_file_line_by_line(file_path):
    print(line)

```

Q8 Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
tuples_list = [(1, 5), (3, 2), (2, 8), (5, 1)]
sorted_list = sorted(tuples_list, key=lambda x: x[1])
sorted_list
```

Q9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
celsius_temp = [25, 30, 35, 40]
fahrenheit_temp = list(map(lambda x: (x * 9/5) + 32, celsius_temp))
fahrenheit_temp
```

Q10 Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
input_string = "Hello, World!"
vowels = ['a', 'e', 'i', 'o', 'u']
filtered_string = ''.join(filter(lambda x: x.lower() not in vowels, input_string))
filtered_string
```
